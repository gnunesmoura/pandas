```python
tickets = pd.read_excel('projeto_0.xlsx', sheet_name='Planilha1')
```

```python
tickets.dtypes
```

```python
def meanOfTicketsOnPeriod(tickets, start, period=1):
    end = start - np.timedelta64(period, 'D')
    return tickets[(tickets['Data Inicial'] <= start) & (tickets['Data Inicial'] > end)]['Duração Real'].mean()
```

```python
period = pd.date_range('20220501', np.datetime64('today'))
rollingAvaregePeriods = [1, 3, 7, 15]
rollingAverage = pd.DataFrame({'SLA': 4}, index=period)

for rollingAvaregePeriod in rollingAvaregePeriods:
    name = 'Média Móvel {} Dias'.format(rollingAvaregePeriod)
    for row in period:
        rollingAverage.at[row, name] = meanOfTicketsOnPeriod(tickets, row, rollingAvaregePeriod)
    rollingAverage[name] = rollingAverage[name] / 60 / 60 / 24
    plt.plot(rollingAverage[name], '*-', label=name)

plt.plot(rollingAverage['SLA'], '*-', label='SLA Atual')

plt.title('SL em Dias', size = 24)
plt.xlabel('Dia da mesuração', size=18)
plt.ylabel('Dias de Trabalho', size=18)
plt.legend()
plt.savefig('figures/sl-dia.png')
plt.show()
```
